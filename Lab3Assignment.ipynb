**Question 1**
(2 points)

Visualize the *Life expectancy at birth, total (years)* of all the countries in the year 2004 in the form of a bar chart (you may drop any null values). Ensure your graph has proper and legible x axis, y axis, labels, and title. Draw a line on the graph that shows the average life expectancy at birth.

```python
cp = df.copy()
cp = cp[df['Year'] == 2004]
cleanCopy = cp.dropna(subset=['Life expectancy at birth, total (years)']) #remove all nulls from the columns

# Extract the required columns
countries = cleanCopy['Country']
life_expectancy = cleanCopy['Life expectancy at birth, total (years)']

#plot a vertical line showing the mean age
meanExp = df['Life expectancy at birth, total (years)'].mean().round(0)

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(countries, life_expectancy, color='blue')
plt.title('Life Expectancy at Birth in 2004 for different Countries')
plt.xlabel('Countries')
plt.ylabel('Life Expectancy at birth (total years)')
plt.axhline(meanExp, color='red', linestyle='dashed', linewidth=2, label='Mean Life Expectancy') #use horizontal line instead of vertical
#horizontal line shows hwo many countries are below the mean and how many are above. Most seem to be above the mean.
#plt.axvline(meanExp, color='red', linestyle='..', label='Mean Life Expectancy')
plt.text(meanExp+0.5, 150, meanExp)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Question 3**
(1.5 point)

Find the relationship between *Internet users (per 1,000 people)* and *Electric power consumption (kWh per capita)* in terms of their correlation. Make sure to state the hypothesis, plot a scatter plot, and discuss it.

Hypothesis:

H1: There is a relationship between Internet users (per 1,000 people) and Electric power consumption (kWh per capita)

H0: There is no relationship between Internet Users (per 1,000 people) and Electric power consumption (kWh per capita)

```python
#Using spearman to test the correlation as both columns can be represented by ratios
from scipy.stats import spearmanr
from matplotlib import pyplot
```

```python
#clean the data
int_users = df.copy().dropna(subset=['Internet users (per 1,000 people)'])
cleanData = int_users.dropna(subset=['Electric power consumption (kWh per capita)'])

# calculate spearman's correlation
coef, p = spearmanr(cleanData['Internet users (per 1,000 people)'], cleanData['Electric power consumption (kWh per capita)'])

#correlation coefficient quantifies the correlation
print('Spearmans correlation coefficient: %.3f' % coef)

# interpret the significance
alpha = 0.05
#p value assesses the significance of correlation
if p > alpha:
 print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
else:
 print('Samples are correlated (reject H0) p=%.3f' % p)
```

The two variables are interdependent or show a positive relationship, according to Spearmans coefficient.

```python
#plot a scatterplot
fig, ax = plt.subplots()
plt.scatter(cleanData['Internet users (per 1,000 people)'], cleanData['Electric power consumption (kWh per capita)'])
ax.set_xlabel("Internet users (per 1,000 people)")
ax.set_ylabel("Electric power consumption (kWh per capita)")
plt.legend()
plt.show()
```

The scatter plot shows a positive linear relationship between the two variables with the exception of some outliers.
